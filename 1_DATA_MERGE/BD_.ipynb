```python
# 1. DATA PROCESSING AND BASE CONSOLIDATION

import os
import pandas as pd

# Direct Raw link to your GitHub repository folder
path = "https://raw.githubusercontent.com/julianevesa/initial-research/main/1_DATA_MERGE/"

# Data Loading
dom1 = pd.read_csv(path + "ACM_SP.csv", encoding="latin1", sep=";").drop(
    columns=["NM_MUN", "NM_UF", "CD_UF"], errors="ignore"
)
dom2 = pd.read_csv(path + "ACM_SP2.csv", encoding="latin1", sep=";").drop(
    columns=["NM_MUN", "NM_UF", "CD_UF"], errors="ignore"
)
dom3 = pd.read_csv(path + "ACM_SP3.csv", encoding="latin1", sep=";").drop(
    columns=["NM_MUN", "NM_UF", "CD_UF"], errors="ignore"
)

basico = pd.read_csv(
    path + "ACM_basico_SP.csv",
    encoding="latin1",
    sep=";",
)

df_favelas = pd.read_csv(path + "pop_favelas_m.csv", sep=";")

# Data Cleaning
columns_to_remove = [
    "CD_REGIAO",
    "NM_REGIAO",
    "CD_UF",
    "NM_UF",
    "CD_NU",
    "NM_NU",
    "CD_AGLOM",
    "NM_AGLOM",
    "CD_RGINT",
    "NM_RGINT",
    "CD_RGI",
    "NM_RGI",
    "CD_CONCURB",
    "NM_CONCURB",
    "AREA_KM2",
    "v0001",
    "v0002",
    "v0003",
    "v0004",
    "v0006",
]
basico = basico.drop(columns=columns_to_remove, errors="ignore")

# Data Standardization
basico["CD_MUN"] = basico["CD_MUN"].astype(str)
dom1["CD_MUN"] = dom1["CD_MUN"].astype(str)
dom2["CD_MUN"] = dom2["CD_MUN"].astype(str)
dom3["CD_MUN"] = dom3["CD_MUN"].astype(str)
df_favelas["COD_MUNICIPIO"] = df_favelas["COD_MUNICIPIO"].astype(str)

# Data Merge
df = (
    basico.merge(dom1, on="CD_MUN", how="left")
    .merge(dom2, on="CD_MUN", how="left")
    .merge(dom3, on="CD_MUN", how="left")
    .merge(df_favelas, left_on="CD_MUN", right_on="COD_MUNICIPIO", how="left")
)

# Post-Merge Adjustments
# Municipalities with no favelas return as NaN; we fill them with 0 and convert to integer
df["N_POP_DPPO"] = df["N_POP_DPPO"].fillna(0).astype(int)

# Remove the duplicated municipality code column from the favelas dataset
df = df.drop(columns=["COD_MUNICIPIO"], errors="ignore")

print(
    f"Final consolidated dataset: {df.shape[0]} municipalities and {df.shape[1]} variables."
)

# Save output locally to Colab environment (Ready to be saved into 2_ANALISE_MODELAGEM later)
df.to_csv("base_final_SP.csv", index=False, sep=";", encoding="utf-8")
print("Successfully generated!")
```
